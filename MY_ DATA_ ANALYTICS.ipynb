```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(my_data['fixed acidity'], my_data['pH'])
plt.xlabel('Fixed Acidity')
plt.ylabel('pH')
plt.title('pH vs Fixed Acidity')
plt.show()
```

**Plot 1: pH vs Fixed Acidity**

The fixed acidity of red wines is plotted versus the pH level in the above scatter plot. Every dot corresponds to a single wine sample. Here is what we see: The scatterplot has a negative gradient and it is evident that there is/would be a stong correlation between the dots and a line of regression drawn through it. There are also a few outliers

```python
plt.hist(my_data['alcohol'], bins=20)
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Content')
plt.show()
```

**Plot 2:Frequency vs Distribution of Alcohol Content**

The alcohol concentration distribution in the red wine dataset is depicted in the above histogram. The frequency of occurrences is shown by the y-axis, while the x-axis shows the amount of alcohol present. Most wines are between X and Y in terms of alcohol percentage, peaking at Z.The highest frequency recorded is 350 while the lowest recorded is close to 0.
